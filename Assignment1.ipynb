Part-2: implementation of bar graph, scatter plot

```python
people=['ann','brandon','chen','david','emily','farook','gagan','lily','imran']
age=[12,20,30,40,50,31,28,15,37]
weight=[55,35,77,68,70,60,72,69,18]
height=[160,135,170,165,173,168,175,159,105]
import matplotlib.pyplot as plt
plt.scatter(weight, height)
plt.title("Relationship between weight nd height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()

plt.figure(figsize=(12,4))
plt.title("people weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people,fontsize=15")
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

```python
#performance of mathematical operation  on the given  datasets multiplication,Division,poweroff, subtraction, modulo2 div
#no operations on list but only on arrayimport numpy as np
dataset1=np.array([4,8,12,24])
dataset2=np.array([0.1,0.01,0.001,0.0001])
product=np.multiply(dataset1,dataset2)
product1=dataset1*dataset2
print("product is as follows",product)
print("another product :",product1)
Division=np.divide(dataset1,dataset2)
Moddivision=np.mod(dataset1,dataset2)
Floordivision=np.floor_divide(dataset1,dataset2)
Subtraction=np.subtract(dataset1,dataset2)
print(Division)
print(Moddivision)
print(Floordivision)
print(Subtraction)
```

```python
#implement trigonometric and log functions of a given data set
import numpy as np
patientage=np.array([65,78,45,67,56,77])
sin1=np.sin(patientage)
cosine=np.cos(patientage)
Tangent=np.tan(patientage)
log1=np.log10(patientage)
log2=np.log2(patientage)
Exp=np.exp(patientage)
print("sine of patient age",sin1)
print("cosine of patient age",cosine)
print("tangent of patient age",Tangent)
print("log of patient age",log1)
print("log base 2 of patient age",log2)
print("exp of patient age",Exp)
```
